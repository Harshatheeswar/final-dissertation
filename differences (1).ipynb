```python
from datasets import Dataset
import re
from transformers import GPT2Tokenizer
from transformers import GPT2Config, GPT2LMHeadModel
import torch
from transformers import DataCollatorForLanguageModeling
from huggingface_hub import notebook_login
from transformers import Trainer, TrainingArguments
import subprocess
```

```python
def preprocess_text(data):
    # Step 1: Remove text within parentheses
    data = re.sub(r'\([^)]*\)', '', data)

    # Step 2: Remove all-uppercase words (likely sound effects)
    data = re.sub(r'\b[A-Z]+\b', '', data)

    # Step 3: Convert to lowercase
    data = data.lower()

    # Step 4: Normalize spaces
    data = re.sub(r'\s+', ' ', data).strip()

    return data

# Load the data
file_path = '/content/drive/MyDrive/data/text_data.zip (Unzipped Files)/train_100M/open_subtitles.train'


with open(file_path, 'r') as file:
    raw_data = file.read()

# Apply preprocessing
cleaned_data = preprocess_text(raw_data)

# Display a sample of the cleaned data
print(cleaned_data[:500])
```

```python
from transformers import AutoTokenizer, AutoConfig, AutoModelForCausalLM

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("babylm/babyllama-100m-2024")

# Load the model configuration without pretrained weights
config1 = AutoConfig.from_pretrained("babylm/babyllama-100m-2024")

model1 = AutoModelForCausalLM.from_config(config1)

config2 = AutoConfig.from_pretrained("Harshatheeswar/gpt2-babytoken")

model2 = AutoModelForCausalLM.from_config(config2)
```

```python
print(config1)
print (config2)
```

```python
max_length = 256
input_texts = [cleaned_data[i:i+max_length] for i in range(0, len(cleaned_data), max_length)]

# Apply the tokenizer to each chunk
tokenized_data = tokenizer(input_texts, padding=True, truncation=True, return_tensors="pt")

# Display a sample of tokenized data
print(tokenized_data["input_ids"][:5])
```

```python
from datasets import Dataset


dataset = Dataset.from_dict(tokenized_data)

# Split the dataset into train and test sets
train_test_split = dataset.train_test_split(test_size=0.1)
train_dataset = train_test_split['train']
test_dataset = train_test_split['test']
```

```python
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
```

```python
# Print the model size
model1_size = sum(t.numel() for t in model1.parameters())
print(f"babyllama size: {model1_size/1000**2:.1f}M parameters")
# Check the maximum sequence length
print(f"Maximum input size for the model: {config1.max_position_embeddings}")
# Print the model2 size
model2_size = sum(t.numel() for t in model2.parameters())
print(f"babyllama size: {model2_size/1000**2:.1f}M parameters")
# Check the maximum sequence length2
print(f"Maximum input size for the model: {config2.max_position_embeddings}")
```

## hidden sizes

512 to 768

```python
from transformers import AutoConfig

desired_hidden_size = 768

# Load the configuration for BabyLLama and modify the hidden size
config1 = AutoConfig.from_pretrained("babylm/babyllama-100m-2024")
config1.hidden_size = desired_hidden_size
config1.intermediate_size = desired_hidden_size * 4


#  modified configs
from transformers import AutoModelForCausalLM
model1 = AutoModelForCausalLM.from_config(config1)
```

```python
model1_size = sum(t.numel() for t in model1.parameters())
print(f"babyllama size: {model1_size/1000**2:.1f}M parameters")
```

```python
config1
```

```python
notebook_login()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model1.to(device)

training_args = TrainingArguments(
    output_dir='Harshatheeswar',
    hub_model_id='babylama-hidden_sizes768',
    evaluation_strategy='epoch',
    auto_find_batch_size=True,
    num_train_epochs=5,
    gradient_accumulation_steps=8,
    weight_decay=0.1,
    lr_scheduler_type='cosine',
    learning_rate=5e-5,
    fp16=True,
    push_to_hub=True,
    logging_steps=10
)

trainer = Trainer(
    model=model1,
    tokenizer=tokenizer,
    data_collator=data_collator,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
)

# Train the model
trainer.train()

# Push the trained model to the Hugging Face Hub
trainer.push_to_hub()
```
